```python
#Import libreries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the data set
df = pd.read_csv("output.csv")
df.head()
```

```python
# Summary statistics
print(df.describe())

```

```python
# Scatter plot: House size vs. Selling price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df)
plt.title('House Size vs. Selling Price')
plt.xlabel('House Size (sqft)')
plt.ylabel('Selling Price ($)')
plt.show()
```

```python
# Box plot: Selling price by number of bedrooms
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Selling Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Selling Price ($)')
plt.show()
```

```python
# Histogram: Distribution of house prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Selling Price ($)')
plt.ylabel('Frequency')
plt.show()
```
